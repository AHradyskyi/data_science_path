```python
import pandas as pd
import numpy as np

#https://github.com/guipsamora/pandas_exercises/blob/master/01_Getting_%26_Knowing_Your_Data/Chipotle/Exercises.ipynb
```

```python
chipo = pd.read_csv("chiptole.csv")
#Import the first 10 rows
chipo.head(10)
```

```python
#The number of observations in the Dataset
rows = chipo.shape[0]
cols = chipo.shape[1]

print("The number of observations are: ", rows)
print("The number  of columns are: ", cols)

#Columns names
columns_name=chipo.columns.tolist()
print("The colummn's names are: ", columns_name)
```

```python
#indexind
ind = pd.DataFrame(chipo)
newdf = ind.set_index('item_name')
print(newdf)
```

```python
#most-ordered item
most_order = chipo.groupby('item_name')
most_order = most_order.sum()
most_order = most_order.sort_values(['quantity'], ascending=False)
most_order.head(1)
```

```python
#most ordered item in the choice_description column
most_order2 = chipo.groupby('choice_description').sum()
most_order2 = most_order2.sort_values(['quantity'], ascending=False)
most_order2.head(1)
```

```python
#Total items sold in total
tota_order2 = chipo.quantity.sum()
print(tota_order2)
```

```python
#Create a lambda function and change the type of item price
def convert_to_float(x):
    try:
        if isinstance(x,str):
            x = x.replace("$","").replace("%","").replace(',', '')
        return(float(x))
    except (ValueError, TypeError):
        return np.nan

chipo['item_price'] = chipo['item_price'].apply(lambda x: convert_to_float(x))

print(chipo['item_price'].dtypes)
```

```python
#Revenue - total amount of money for all sold items(
revenue = (chipo['quantity']*chipo['item_price']).sum()
print(revenue)
```

```python
orders = chipo.order_id.value_counts().count()
print("Number of orders",orders)
```

```python
#Average price in the revenue
av_p = chipo.groupby('order_id')['revenue'].sum().mean()
print(np.round(av_p,2))
```

```python
item_sold = chipo.item_name.value_counts().count()
print(item_sold)
```
